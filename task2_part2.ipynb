```python
import pandas as pd

# Load only 100,000 rows
data = pd.read_csv('flights_cleaned.csv', nrows=100000)

print(data.shape)
data.head()

```

### Data Cleaning & Feature Prep

#### 1️⃣ Drop extra columns

```python
# Drop duplicate or unneeded merge columns
data.drop(['IATA_CODE', 'IATA_CODE_ORIGIN'], axis=1, inplace=True, errors='ignore')
```

#### 2️⃣ Select features



```python
features = ["AIRLINE_CODE", "ORIGIN_CITY", "ORIGIN_STATE"]
X = data[features]
y = data["DELAYED"]
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
print(X_encoded.head())
```

#### 4️⃣ Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

### Build & Train Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

#### 2️⃣ Fit the model

```python
# Train on training data
model.fit(X_train, y_train)
```

#### 3️⃣ Make predictions

```python
# Predict on test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

####  4️⃣ Evaluate performance

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

#### 5️⃣ (Optional) Check predicted probabilities

```python
# Predict probabilities for ROC curve later (optional)
y_prob = model.predict_proba(X_test)[:, 1]
```
